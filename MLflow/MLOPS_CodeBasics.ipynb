```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
import warnings

warnings.filterwarnings('ignore')
```

```python
X,y = make_classification(n_samples=1000,n_features=10,n_informative=2,n_redundant=8,weights=[0.9,0.1],flip_y=0,random_state=42)

np.unique(y,return_counts=True)
```

```python
X.shape
```

```python
y[:4]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
params = {
    "solver":"lbfgs",
    "max_iter":1000,
    "multi_class":"auto",
    "random_state":8888
}

lr = LogisticRegression(**params)

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

```

```python
y_pred, y_test
```

```python
report = classification_report(y_test,y_pred)
print(report)
```

```python
rep_dict = classification_report(y_test,y_pred,output_dict=True)
rep_dict
```

```python
rep_dict['macro avg']['f1-score']
```

```python
import mlflow
```

```python
mlflow.set_experiment('Venky Experiment')

mlflow.set_tracking_uri("http://127.0.0.1:5000/")

with mlflow.start_run():
    mlflow.log_params(params)
    mlflow.log_metrics(
        {
            'accuracy':rep_dict['accuracy'],
            'recall_class_0': rep_dict['0']['recall'],
            'recall_class_1': rep_dict['1']['recall'],
            'f1_score_macro_avg':rep_dict['macro avg']['f1-score']            
        }
    )
    mlflow.sklearn.log_model(lr,'Logistic Regression')
```

# 1. Logistic Regression

```python
log_Reg = LogisticRegression(C=1,solver='liblinear')
log_Reg.fit(X_train,y_train)
y_pred_log_reg = log_Reg.predict(X_test)
print(y_pred_log_reg[:5])
print(y_test[:5])
print(classification_report(y_test,y_pred_log_reg))
```

# 2. Random Forest

```python
rfc = RandomForestClassifier(n_estimators=30,criterion='gini',max_depth=3)
rfc.fit(X_train,y_train)
y_pred_rfc = rfc.predict(X_test)
print(y_pred_rfc[:5])
print(y_test[:5])
print(classification_report(y_test,y_pred_rfc))
```

# 3. XGBoost

```python
xgb = XGBClassifier(use_label_encoder = False,eval_metric = 'logloss')
xgb.fit(X_train,y_train)
y_pred_xgb = xgb.predict(X_test)
print(y_pred_xgb[:5])
print(y_test[:5])
print(classification_report(y_test,y_pred_xgb))
```

# 4. Handle Class imbalance using SMOTETomek and then train using XGBoost

```python
from imblearn.combine import SMOTETomek

smt = SMOTETomek(random_state=42)
X_train_res,y_train_res = smt.fit_resample(X_train,y_train)

np.unique(y_train_res,return_counts=True)
```

```python
xgb_cf = XGBClassifier(use_label_encoder=False,eval_metric='logloss')
xgb_cf.fit(X_train_res,y_train_res)
y_pred_xgb_cf = xgb_cf.predict(X_test)
print(classification_report(y_test,y_pred_xgb_cf))
```

```python
reports =[]
for model_name,params,model,train_data,test_data in models:
    X_train = train_data[0]
    y_train = train_data[1]
    X_test = test_data[0]
    y_test = test_data[1]

    model.set_params(**params)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test,y_pred,output_dict=True)
    reports.append(report)
    
    
```
